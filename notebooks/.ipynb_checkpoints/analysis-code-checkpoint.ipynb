```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
from scipy.stats import chi2_contingency
```

```python
# Read & clean layoffs.csv data

data = pd.read_csv('/Users/jamesfisher/Documents/GitHub/Data-Science-Project/data/layoffs.csv')
# display(data.head())
data['date'] = pd.to_datetime(data['date'])
# print(data['date'].dtype)

data['location'] = data['location'].fillna('Unknown')
data['industry'] = data['industry'].fillna('Unknown')
data['stage'] = data['stage'].fillna('Unknown')
# Replaced missing total_laid_off values with 0, assuming no layoffs were reported for these rows
data['total_laid_off'] = data['total_laid_off'].fillna(0)
# print(data.isnull().sum())
```

```python
# What trends can be observed in the number of layoffs over time?

# 1. Ensure all dates are in datetime64
print(data['date'].dtype)

# 2. Group the data by month/year & calculate the total number of layoffs for each time period
monthly_layoffs = data.groupby(data['date'].dt.to_period('M'))['total_laid_off'].sum()
# print(monthly_layoffs)

# 3. Compute a moving average
monthly_layoffs_moving_avg = monthly_layoffs.rolling(window=3).mean()

# 4. Plot the raw aggregated data and the moving average to visualize trends over time
plt.figure(figsize=(12, 6))
plt.plot(monthly_layoffs.index.astype(str), monthly_layoffs, label='Monthly Layoffs', linewidth=2)
plt.plot(monthly_layoffs.index.astype(str), monthly_layoffs_moving_avg, label='3-Month Moving Avg', linewidth=2, linestyle='--')
plt.title('Trends in Tech Layoffs (2020-2024)', fontsize=16)
plt.xlabel('Time (Month)', fontsize=12)
plt.ylabel('Total Layoffs', fontsize=12)
plt.legend()
plt.xticks(rotation=80)
plt.tight_layout()
plt.show()
```

```python
# Are certain industries or sectors within tech more prone to layoffs during economic downturns?

# 1. Group the data by industry and calculate the total number of layoffs for each sector
industry_layoffs = data.groupby('industry')['total_laid_off'].sum()
print(industry_layoffs)

# 2. Visualize the total layoffs by industry with a bar plot
plt.figure(figsize=(12, 6))
industry_layoffs.sort_values(ascending=False).plot(kind='bar', color='skyblue')
plt.title('Total Layoffs by Industry (2020-2024)', fontsize=16)
plt.xlabel('Industry', fontsize=12)
plt.ylabel('Total Layoffs', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Were there any periods of heightened layoffs that align with major economic or global events (e.g., pandemic waves, inflation spikes)?

# 1. Define key events and their dates
events = {
    "COVID-19 Pandemic Start": "2020-03",
    "Inflation Spike": "2022-11"
}

# 2. Aggregate data for each event
simplified_event_analysis = {}
for event, date in events.items():
    event_date = pd.Period(date, freq='M')
    
    # Calculate layoffs for pre-event, during-event, and post-event periods
    pre_event = monthly_layoffs.loc[event_date - 3:event_date - 1].sum()
    during_event = monthly_layoffs.loc[event_date:event_date].sum()
    post_event = monthly_layoffs.loc[event_date + 1:event_date + 3].sum()
    simplified_event_analysis[event] = [pre_event, during_event, post_event]

# 3. Create a visualization
simplified_event_df = pd.DataFrame(simplified_event_analysis, index=["Pre-Event", "During Event", "Post-Event"])
simplified_event_df.T.plot(kind='bar', figsize=(10, 6), color=['#4caf50', '#2196f3', '#f44336'])
plt.title("Layoffs During Major Events (COVID-19 and Inflation Spike)", fontsize=16)
plt.ylabel("Total Layoffs", fontsize=12)
plt.xlabel("Event", fontsize=12)
plt.xticks(rotation=0)
plt.legend(title="Time Period")
plt.tight_layout()
plt.show()
```
